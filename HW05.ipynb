# Q1. 寫一個函式，抓取 Google Finance 的股價歷史資料

```python
import pandas as pd
import io
import requests 
```

```python
url = "http://finance.google.com/finance/historical"

def google_stock(stock,startdate,enddate):    
    params = {}
    params['q'] = stock
    params['startdate'] = startdate
    params['enddate'] = enddate
    params['output'] = 'csv'
    
    r = requests.get(url,params=params)
    stockdata = pd.read_csv(io.StringIO(r.text))
    
    return(stockdata)
```

```python
google_stock(stock='AAPL', startdate='2016-01-01', enddate='2016-01-31')
```
